# PYTHON SHOPPING CART

### Build a shopping cart program with the following capabilities:

##### 1) Takes in an input
##### 2) Stores user input into a dictionary or list
##### 3) The User can add or delete items
##### 4) The User can see current shopping list
##### 5) The program Loops until user 'quits'
##### 6) Upon quiting the program, prints out a receipt of the items with total and quantity.

```python

## ADD Function ##

# declare a function that will add items to the cart dictionary
# ask user for input regarding the item, price & quantity
# if the item is in the cart update it in cart, 
# if not add the item as key 
    # with value being another dictionary with  price and quantity as the key and corresponding user input


def add_decision():
    item = input("What are you adding to your cart?: ")
    price = float(input("What is the price?: "))
    quantity = int(input("How many?: "))
    if item in cart:
        cart[item]["price"] = price
        cart[item]["quantity"] += quantity
    else:
        cart[item] = {"price": price, "quantity":quantity}
    print(f"{quantity} {item} have been added to your cart!")
    

# ==============================================================================    


## DELETE Function ##

# declare a function that will delete items from the cart dictionary
# ask user for input regarding what item to delete
# remove item from dictionary


# def delete_decision():
#     item_to_delete = input("What item would you like to delete?: ")
#     if item_to_delete in cart:
#         cart.pop(item_to_delete)
#         print(f"{item_to_delete} has been removed from cart") 
#     else:
#         print(f"{item_to_delete} was not found in cart")
        

def delete_decision():
    item = input("What item would you like to delete?: ")
    quantity = int(input("How many would you like to delete?: "))
    if item in cart:
        cart[item]["quantity"] -= quantity
        print(f"{quantity} {item} has been deleted from your cart")
        if cart[item]["quantity"] <= 0:
            del cart[item]
            print(f"{item} has been removed from your cart")
    else:
        print(f"{item} was not found in your cart")

# ==============================================================================           


## SHOW Function ##

# iterate through cart dictionary to access keys and values
# print out the item, price and quantity
    
    
def show_decision():
    print("="*30)
    print("\nItems Currently In Your Cart:")
    for item, data in cart.items():
        print(f"\n{item}: ${data['price']:.2f} x {data['quantity']}")
    print("="*30)    

# ==============================================================================       


## RECIEPT Function ##

# declare a variable to store the total price of items in the cart
# iterate through the cart to access the keys and values
# set variables that will store the values of the price and quantity key 
# total variable will be the product of  items price and item quantity
# print total

    
def reciept():
    total = 0
    print('='*20)
    print('\nHere is your reciept\n\n')
    for item, data in cart.items():
        item_price = data["price"]
        item_quantity = data["quantity"]
        print(f"{item_quantity} {item} at ${item_price:.2f} Each\n")
        total += item_price * item_quantity
    print("~"*20)
    print(f"\nYour Total Is: ${total:.2f}")
    print("\n\nThank you, come Again!\n")
    print("-"*20)

    
```

```python
# declare a function that will be the main function that runs the other functions
# create a while-loop to run until user 'quits'
# get user input on what action to select for the shopping cart
    # what ever selection is chosen should run the corresponding function for it
# if user selection is not a valid selection; print error message



cart = {}


def shopping_cart():
    
    while True:
        choice = input("What would you like to do? add|delete|show|quit:  ").lower()
        if choice == 'add':
            add_decision()
        elif choice == 'delete':
            delete_decision()
        elif choice == 'show':
            show_decision()
        elif choice == 'quit':
            reciept()
            break
        else:
            print("please enter a valid choice!".title())
        

#call the shopping cart function    
shopping_cart()
```
